```python
import numpy as np
import matplotlib.pyplot as plt
import math

n = 20
d=3
A = np.random.randn(n, d)
A /= np.sqrt(np.sum(A**2, axis=1)).reshape(-1, 1) 
B = 1 + np.random.random(n)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)

def d(a,b):
    return math.sqrt((a[0]-b[0])**2+(a[1]-b[1])**2+(a[2]-b[2])**2)

x0=np.zeros((n,n,3))
for a in range(n):
    for b in range(a+1,n):
        for c in range(b+1,n):
            AA=[A[a],
               A[b],
               A[c]]
            BB=[B[a],
                B[b],
                B[c]]   
            if (np.linalg.det(AA)!=0):
                x = np.linalg.solve(AA, BB)
            for i in range(n):
                if (A[i][0]*x[0]+A[i][1]*x[1]+A[i][2]*x[2]>B[i]):
                    break
                if(i==n-1):
                    ax.scatter(x[0],x[1],x[2])
                    if(x0[a][b][0]==x0[a][b][1]==x0[a][b][2]==0):
                        x0[a][b]=x
                    else:
                        print(x0[a][b])
                        ax.plot([x0[a][b][0],x[0]], [x0[a][b][1],x[1]], [x0[a][b][2],x[2]])
    
plt.show()
```
