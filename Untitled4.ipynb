1. for every three inequalities we can find a possible point intersection, so to find all the set, it takes n^3. Then we want to make sure every point satisfy all the inequalities, so we take the value of point to each inequality to compare which takes n times. So in total, n^4 times.

2. for every intersection of two faces, or two constraints, there is a line which contains two vertexs, then it is an edge. 

3. since we find out all the vertexs, all the constraints that makes the intersection are not redundant. Besides that, other constraints are redundant as they didn't construct the point. 

4. we know a face is defined by a constraint, then for all non redundant face, there should be at least three points that defined it and they can form an active constraint corresponding to F. 

```python
import numpy as np
import matplotlib.pyplot as plt
import math

n = 20
d=3
A = np.random.randn(n, d)
A /= np.sqrt(np.sum(A**2, axis=1)).reshape(-1, 1) 
B = 1 + np.random.random(n)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)

def d(a,b):
    return math.sqrt((a[0]-b[0])**2+(a[1]-b[1])**2+(a[2]-b[2])**2)

x0=np.zeros((n,n,3))
for a in range(n):
    for b in range(n):
        for c in range(n):
            AA=[A[a],
               A[b],
               A[c]]
            BB=[B[a],
                B[b],
                B[c]]   
            if (np.linalg.det(AA)!=0):
                x = np.linalg.solve(AA, BB)
                for i in range(n):
                    if (A[i][0]*x[0]+A[i][1]*x[1]+A[i][2]*x[2]>B[i]):
                        break
                    if(i==n-1):
                        ax.scatter(x[0],x[1],x[2])
                        if(x0[a][b][0]==x0[a][b][1]==x0[a][b][2]==0):
                            x0[a][b]=x
                        else:
                            ax.plot([x0[a][b][0],x[0]], [x0[a][b][1],x[1]], [x0[a][b][2],x[2]])
    
plt.show()
```

```python
n = 20
d=2
A = np.random.randn(n, d)
A /= np.sqrt(np.sum(A**2, axis=1)).reshape(-1, 1) 
B = 1 + np.random.random(n)
c = np.random.randn(2)

plt.figure(figsize = (5, 5))

for a in range(n):
    for b in range(a+1,n):
        AA=[A[a],
           A[b]]
        BB=[B[a],
            B[b]]   
        if (np.linalg.det(AA)!=0):
            x = np.linalg.solve(AA, BB)
        for i in range(n):                
            if (A[i][0]*x[0]+A[i][1]*x[1]>B[i]):
                break
            if(i==n-1):
                plt.scatter(x[0],x[1])
                xx = np.linspace(-2,2,100)
                yy1 = (B[a]-A[a][0]*xx)/A[a][1]
                yy2 = (B[b]-A[b][0]*xx)/A[b][1]
                plt.plot(xx, yy1, '-r')
                plt.plot(xx, yy2, '-r')

plt.xlim(-3, 3)
plt.ylim(-3, 3)          
xx, yy = np.meshgrid( np.linspace(-3, 3, 1000), np.linspace(-3, 3, 1000) )
z = c[0]*xx+c[1]*yy
im1 = plt.contour( xx, yy,z, levels = 20) 
plt.colorbar(im1)
```
